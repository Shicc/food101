```python
# import h5py
import os
import matplotlib.pyplot as plt
from keras.utils.io_utils import HDF5Matrix

train_h5_path = os.path.join('food_c101_n1000_r384x384x3.h5')

images = HDF5Matrix(train_h5_path, 'images',start=0, end=10)
print('images.shape:',images.shape)
category_names = HDF5Matrix(train_h5_path,'category_names',start=0, end=10)
print('category_names.shape:',category_names.shape)
%matplotlib inline
```

```python
for i in range(10):
    print(i)
    plt.imshow(images[i].reshape(384,384,3),cmap='gray')
    plt.show()
```

```python
for i in range(10):
    print(i,category_names[i])
```
